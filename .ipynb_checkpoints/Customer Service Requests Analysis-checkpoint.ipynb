```python
import warnings
warnings.filterwarnings('ignore')
import os
os.chdir('D:/ALL PYTHON/SimplilearnProject/Data-Science-with-Python-Project-2--master/')
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import math
import dateutil
import datetime as dt
```

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols 
```

```python
##reading the file 
complain_data = pd.read_csv('311_Service_Requests_from_2010_to_Present.csv')
complain_data.head()
```

```python
##getting the structural information
print('The shape of the data frame is {}'.format(complain_data.shape))
print('The comulmn names are {}'.format(complain_data.columns))
print('The row names are {}'.format(complain_data.index))
```

```python
##checking the data type 
print('The data type of each variable is\n{}'.format(complain_data.dtypes))
```

```python
##checking for null values 
#print('Data contains null value in the following columns\n{}'.format(complain_data.isna().sum()))
##checking the null values using the heat map
plt.subplots(figsize = (30,10))
sns.heatmap(complain_data.isna(),yticklabels=False)
plt.show()
```

```python
# Read or convert the columns ‘Created Date’ and Closed Date’ to datetime datatype
#and create a new column ‘Request_Closing_Time’ as the time elapsed between request creation and request closing. 
complain_data[['Created Date','Closed Date']] = complain_data[['Created Date','Closed Date']].apply(pd.to_datetime)
```

```python
complain_data['Request_Closing_Time'] = complain_data['Closed Date'] - complain_data['Created Date']
complain_data.head(3)
```

```python
##lets find the average hours in which the complaint was closed
print('The average time taken to resolve the complaints is {} '.format(complain_data.Request_Closing_Time.mean()))
```

```python
##lets make a column which will give the time difference in minutes.

complain_data['TimeTakenInMinutes'] =  round(complain_data.Request_Closing_Time.dt.total_seconds()/60,3)
complain_data.head(3)
```

#### Analysis of Complaint Type

```python
##lets make the table by grooping the complaint type  and mean of TimeTakenInMinutes
time_complaint_type = complain_data.groupby('Complaint Type')['TimeTakenInMinutes'].mean().reset_index()
time_complaint_type.sort_values(by = 'TimeTakenInMinutes',ascending = False,inplace = True)
time_complaint_type.head(6)
```

```python
fastest_comp = time_complaint_type.loc[time_complaint_type['TimeTakenInMinutes'] == time_complaint_type.TimeTakenInMinutes.min()]
slowest_comp = time_complaint_type.loc[time_complaint_type['TimeTakenInMinutes'] == time_complaint_type.TimeTakenInMinutes.max()]
```

```python
print("The complain type which is resolved fastest is '{}' with the average time of {} minutes"\
      .format(fastest_comp.loc[19,'Complaint Type'],round(fastest_comp.loc[19,'TimeTakenInMinutes'],3)))
print("The complaint type '{}' takes maximum time of {} minutes on average to get resolved"\
      .format(slowest_comp.loc[2,'Complaint Type'],round(slowest_comp.loc[2,'TimeTakenInMinutes'],3)))
```

```python
##whats happening with Ferry Complaints
##closing date either not mentioned or the complaints are still pending
ferry_comp = complain_data[complain_data['Complaint Type'] == 'Ferry Complaint']
ferry_comp
```

```python
##since 'Animal in a Park' complaint type has taken way more time to resolve, thats why dropping that record
##to maintain the uniformity of the bar graph
time_complaint_type.drop(2,inplace = True)
```

```python
##dropping 'Ferry Complaint' complaint type since the closed date is not mentioned, or it has not been solved yes
time_complaint_type.drop(8,inplace=True)
```

```python
##plotting the graphl of complaint type and average time taken
plt.figure(figsize=(15,5))
comp_type_plt = sns.barplot(data = time_complaint_type,x = 'Complaint Type', y = 'TimeTakenInMinutes',palette = 'Greens_d')
plt.xticks(rotation = 70)
plt.title('Mean Time Taken To Resolve Various Complaints\n')
plt.ylabel('Average Time Taken To Resolve\n in (minutes)')
plt.show()
```

```python
##finding out time taken by complains according to location type
complain_data['Location Type'].unique()
```

```python
location_data = complain_data.groupby('Location Type')['TimeTakenInMinutes'].mean().reset_index()
location_data.sort_values('TimeTakenInMinutes',ascending = False, inplace = True)
location_data
```

```python
print("The location type which is resolved fastest is '{}' with the average time of {} minutes"\
      .format(location_data.loc[7,'Location Type'],round(location_data.loc[7,'TimeTakenInMinutes'],3)))
print("The location type '{}' takes maximum time of {} minutes on average to get resolved"\
      .format(location_data.loc[15,'Location Type'],round(location_data.loc[15,'TimeTakenInMinutes'],3)))
```

```python
##whats happening with the complaints comming from terminal
complain_data.loc[complain_data['Location Type'] == 'Terminal',:]
```

```python
##since the the complaints from park are taking way more time to get resolve, and compaints of Ferry and terminal are not solved
##so dropping those record to get a uniform bar graph
location_data.drop([7,3,16],inplace=True)
location_data.head(3)
```

```python
##plotting the grapg
plt.figure(figsize=(15,7))
sns.barplot(x = 'Location Type', y = 'TimeTakenInMinutes',data = location_data,palette = 'Reds_r')
plt.xticks(rotation = 75)
plt.title('Average Time Taken By Complaints From Different Area\n')
plt.ylabel('Time Taken\n in (minutes)')
plt.show()
```

```python
complain_data.City.unique()
```

```python
##lets group by and plot the average closing time according to city
city_data = complain_data.groupby('City')['TimeTakenInMinutes'].mean().reset_index()
city_data.sort_values('TimeTakenInMinutes',ascending = False,inplace = True)
city_data.head(4)
```

```python
#plotting graphs of countries
#pal1 = sns.color_palette("flare", as_cmap=True)
#pal2 = sns.color_palette("rocket", as_cmap=True)
fig,ax = plt.subplots(1,2,figsize = (20,10))
sns.barplot(x = 'TimeTakenInMinutes',y = 'City',data = city_data[:10],ax = ax[0],palette = 'twilight')
sns.barplot(x = 'TimeTakenInMinutes',y = 'City',data = city_data[-10:],ax = ax[1],palette= 'ocean_r')
for i in range(2):
    ax[i].set_xlabel('Time in Minutes',size = 15)
    if i == 0:
        ax[i].set_title('Top 10 Slowest Acting Cities\n',size = 22)
        ax[i].set_ylabel('City (SLOWEST)',size = 15)
    else:
        ax[i].set_ylabel('City (FASTEST)',size = 15)
        ax[i].set_title('Top 10 Fastest Acting Cities\n',size = 22)
    
fig.tight_layout()
```

```python
##checking if the distribution is normal or not 
mod = ols('Complaint Type ~ TimeTakenInMinutes', data  = complain_data).fit()
anov_table = sm.stats.anova_lm(mod, type = 2)
print(anov_table)
```
